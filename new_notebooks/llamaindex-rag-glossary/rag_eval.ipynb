```python
RAG_OUTPUTS = "improved_rag_results.json"
```

```python
import pandas as pd
df = pd.read_json(RAG_OUTPUTS)
```

```python
# INSERT YOUR OPENAI API KEY HERE!!!
import os
os.environ["OPENAI_API_KEY"] = #...
```

```python
from datasets import Dataset

dataset = Dataset.from_pandas(df)
```

```python
from ragas.metrics import (
    faithfulness,
    answer_similarity,
    answer_relevancy,
    answer_correctness,
    context_precision,
    context_recall,
    context_relevancy,
)
from ragas import evaluate

eval_results = evaluate(
    dataset=dataset,
    metrics=[
        faithfulness,
        answer_similarity,
        answer_relevancy,
        answer_correctness,
        context_precision,
        context_recall,
        context_relevancy,
    ]
)
```

```python
eval_results
```

```python
eval_df = eval_results.to_pandas()
```
